# Pre-Processing of Text
Summary: Taking the text of a single column and removing stop-words,
punctuation, and getting word counts

```python
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile

df = pd.read_excel('Acute_Threat_Fear.xlsx', sheet_name = 'Sheet1')
```

```python
from nltk.tokenize import word_tokenize

# list of all words, punctuation, stop words
list_strings = [word_tokenize(i) for i in df['abstract']] 
```

```python
from itertools import chain
import string

#Conversion from 'list of lists' -> string
#Further cleaning from NLTK
output = list(chain(*list_strings))
corpus = ' '.join(str(word) for word in output)
corpus = corpus.lower().translate(str.maketrans('', '', string.punctuation))
corpus
#Bug: Combines words together if punctuation is within word
#ex. war-related -> warrelated 
```

```python
#Function that will count number of words

def word_count(str):
    counts = dict()
    words = str.split()
    
    for word in words:
        if word in counts:
            counts[word] += 1
        else:
            counts[word] = 1
            
    return counts
```

```python
#Function that will print descending counts of words in corpus

from collections import OrderedDict

def count_corpus(corpus):
    num = word_count(corpus)
    num_keys = sorted(num, key = num.get, reverse = True)

    #We can see counts of each word & possibly other stop words to add
    for word in num_keys:
        print(word, num[word])
        
```

# Word Counts (uncleaned)

```python
count_corpus(corpus)
```

```python
from nltk.corpus import stopwords

#Removing all stop words from NLTK word list
stop = list(stopwords.words('english') )

corpus_split = corpus.split()
corpus_cleaned = [word for word in corpus_split if word.lower() not in stop]
corpus_cleaned = ' '.join(corpus_cleaned)
corpus_cleaned
```

# Word Counts (cleaned)

```python
count_corpus(corpus_cleaned)
```

Based on the top words, there definitely seems to be some value since it contains unique words such as: threat, fear, acute, amygdala, etc..

# N-Gram Function 
Used to explore the dataset a little more (bi, tri, etc..) 


```python
from nltk import ngrams

def ngrams_funct(n, corpus):
    
    bigrams = ngrams(corpus.split(), n)
    freq_bigrams = nltk.FreqDist(bigrams)

    return freq_bigrams.most_common(10)
```

```python
bigrams = ngrams_funct(2, corpus_cleaned)
bigrams
```

```python
trigrams = ngrams_funct(3, corpus_cleaned)
trigrams
```

# Stemming Corpus 
Afterwards checking the word count to see if anything dramatic has changed within Top 5 Words

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

#Porter Stemmer is known to be a cruel way of stemming
ps = PorterStemmer()

stemmed_list = [ps.stem(word) for word in corpus_split]
stemmed_list
```

```python
stemmed_corpus = [word for word in stemmed_list if word.lower() not in stop]
stemmed_corpus = ' '.join(stemmed_corpus)
count_corpus(stemmed_corpus)
```

# Lemmatization

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
lemmatized_list = [lemmatizer.lemmatize(word) for word in corpus_split]

lemmatized_list
```

```python
lemmatized_corpus = [word for word in lemmatized_list if word.lower() not in stop]
lemmatized_corpus = ' '.join(lemmatized_corpus)
count_corpus(lemmatized_corpus)
```
